```python
def scanner(urls):
  output = []
  with Driver(uc=True, uc_cdp_events=True) as driver:
    raw = []
    # Handle traffic data
    def save(data):
        req = {}
        req["id"] = data["params"]["requestId"]
        req["path"] = data["params"]["headers"][":path"]
        if "/getData/" in req["path"]:
            raw.append(req)
    # Listen traffic
    driver.add_cdp_listener('Network.requestWillBeSentExtraInfo', save)

    # Foreach urls
    for url in urls:
      # Get page
      driver.open(url)
      print(f"Open {url}")
      driver.sleep(3)

      # Close cookie notice
      try:
        driver.click('button:contains("AGREE")', timeout=3)
      except Exception as e:
         pass

      driver.sleep(1)

      print(f"Zoom in / out")
      # Zoom in
      for i in range(3):
          driver.click('a[title="Zoom in"]')
          driver.sleep(1)

      # Zoom out
      for i in range(2):
          driver.click('a[title="Zoom out"]')
          driver.sleep(1)

      driver.click("div#map_canvas")
      map = driver.find_element("div#map_canvas")

      offset_x = 0
      offset_y = 0

      # MV = 10
      # CR = 10

      # print(f"Area scan")
      # # Area scan
      # for i in range(MV):
      #     # Forward
      #     for j in range(MV):
      #       map.send_keys(Keys.ARROW_RIGHT)
      #       offset_x += 1
      #       driver.sleep(0.4)
      #     # Carret
      #     for j in range(CR):
      #       map.send_keys(Keys.ARROW_DOWN)
      #       offset_y -= 1
      #       driver.sleep(0.2)
      #     # Backward
      #     for j in range(MV):
      #       map.send_keys(Keys.ARROW_LEFT)
      #       offset_x -= 1
      #       driver.sleep(0.4)
      #     # Carret
      #     for j in range(CR):
      #       map.send_keys(Keys.ARROW_DOWN)
      #       offset_y -= 1
      #       driver.sleep(0.2)

      # Area scan
      print(f"Area scan")
      MV = 5
      # Right
      for i in range(MV):
        map.send_keys(Keys.ARROW_RIGHT)
        offset_x += 1
        driver.sleep(0.4)
      # Left
      for i in range(MV * 2):
        map.send_keys(Keys.ARROW_LEFT)
        offset_x += 1
        driver.sleep(0.4)
      # Up
      for i in range(MV):
        map.send_keys(Keys.ARROW_UP)
        offset_y += 1
        driver.sleep(0.4)
      # Down
      for i in range(MV * 2):
        map.send_keys(Keys.ARROW_DOWN)
        offset_y += 1
        driver.sleep(0.4)

      print(f"Back to center")
      # Back to center
      while offset_x != 0:
        if offset_x > 0:
          offset_x -= 1
          map.send_keys(Keys.ARROW_RIGHT)
          driver.sleep(0.2)
        if offset_x < 0:
          offset_x += 1
          map.send_keys(Keys.ARROW_LEFT)
          driver.sleep(0.2)
      while offset_y != 0:
        if offset_y > 0:
          offset_y -= 1
          map.send_keys(Keys.ARROW_DOWN)
          driver.sleep(0.2)
        if offset_y < 0:
          offset_y += 1
          map.send_keys(Keys.ARROW_UP)
          driver.sleep(0.2)

      # print(f"Zoom in")
      # # Zoom in
      # for i in range(3):
      #     driver.click('a[title="Zoom in"]')
      #     driver.sleep(1)

      print(f"Zoom out")
      # Zoom out
      for i in range(5):
          driver.click('a[title="Zoom out"]')
          driver.sleep(1)

      driver.sleep(1)

      # Handle all traffic
      print(f"Handle all traffic")
      for req in raw:
          try:
              response = driver.execute_cdp_cmd(
                  "Network.getResponseBody", {"requestId": req["id"]})
              out = {"response": response} | req
              # Save responses
              output.append(out)
          except Exception as e:
              pass
      print(f"Handle traffic done!")

      driver.sleep(3)
      raw.clear()

    # Close browser
    driver.close()
  return output
```

```python
def shipRawParser(raw):
    regex = r"z:(?P<z>\d+)\/X:(?P<x>\d+)\/Y:(?P<y>\d+)"

    shipData = []
    for r in raw:
        if "get_data_json_4" in r["path"]:
            matches = re.findall(regex, r["path"])
            z = matches[0][0]
            x = matches[0][1]
            y = matches[0][2]

            rows = json.loads(r["response"]["body"])["data"]["rows"]
            for row in rows:
                data = {}
                data["z"] = z
                data["x"] = x
                data["y"] = y

                data["data"] = row

                shipData.append(data)

    return shipData

def shipDataParser(shipData):
  ships = {}

  for data in shipData:
    ship = {} | data["data"]
    ship["TILE_X"] = data["x"]
    ship["TILE_Y"] = data["y"]
    ship["TILE_Z"] = data["z"]

    id = data["data"]["SHIP_ID"]

    # Keep ship with highest Z value
    if id in ships:
      if int(ship["TILE_Z"]) > int(ships[id]["TILE_Z"]):
        ships[id] = ship
    else:
      ships[id] = ship

  return ships
```
